```python
##import libraries
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

```python
CovidRawData = pd.read_json("https://covid.ourworldindata.org/data/owid-covid-data.json")
CovidRawData
```

```python
CovidRawData_transposed = CovidRawData.T

dataList = CovidRawData_transposed['data'][0]
dataList[0]
```

```python
x_values = []
for index in range(len(dataList)):
    for key in dataList[index]:
        if key == 'date':
            x_values.append(dataList[index][key])
x_values = x_values[0:236]
```

```python
y_values = []
for index in range(len(dataList)):
    for key in dataList[index]:
        if key =='total_cases_per_million':
            y_values.append(dataList[index][key])
len(y_values)
```

```python
plt.scatter(x_values, y_values, marker="o", facecolors="red", edgecolors="black", alpha=0.75)
```
